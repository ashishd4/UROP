```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
```

```python
Lmin = 10
Lmax = 100
steps = 19
c = 1
start = 2
end = 5
height = 5
x = np.linspace(Lmin, Lmax, steps)

for i in range(5):
    vals = np.linspace(1, 1, steps)
    vals[c + start:c + end] = height
    plt.plot(x, vals)
    plt.grid()
    plt.show()
    
    c += 2
```
