```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import LineString
```

# Riemann Solver Exercise with 2 Cell Interfaces

```python
# Ideal gas constant for air
R = 287 # J/g*K

# Set initial values for each cell
u0 = 600 # m/s
T0 = 300 # K
P0 = 100 # kPa
rho0 = P0/(R*T0)

u1 = 500 # m/s
T1 = 200 # K
P1 = 200 # kPa
rho1 = P1/(R*T1)

u2 = 400 # m/s
T2 = 100 # K
P2 = 300 # kPa
rho2 = P2/(R*T2)

# Set initial values at the interfaces with the cell values
uL = (u0 + u1)/2
TL = (T0 + T1)/2
PL = (P0 + P1)/2
rhoL = (rho0 + rho1)/2

uR = (u1 + u2)/2
TR = (T1 + T2)/2
PR = (P1 + P2)/2
rhoR = (rho1 + rho2)/2

# Other constants
dx = 0.1 # m
CFL = 0.5
gamma = 1.4 # for air
cp = 1.005 # J/g*K
cv = 0.718 # J/g*K
```

```python
P_arr = [[P0, P1, P2]]
u_arr = [[u0, u1, u2]]
T_arr = [[T0, T1, T2]]
rho_arr = [[rho0, rho1, rho2]]
t_arr = [0]
```

```python
# Initial and populate time array
t = np.arange(1, 51) * 0.000001 - 0.000001
    
# Initial and populate space array
x = np.arange(1, 51) * 0.000001

# Speed of sound variables
aL = np.sqrt(gamma*R*TL)
aR = np.sqrt(gamma*R*TR)

# Defining characteristic lines
x1L = (uL + 2*aL/(gamma - 1)) * t
x2L = (uL - 2*aL/(gamma - 1)) * t

x1R = (uR + 2*aR/(gamma - 1)) * t + dx
x2R = (uR - 2*aR/(gamma - 1)) * t + dx

# Solving for intersection time, and multiplying by the CFL number
intersection_time = dx/((uL - uR) + 2*(aL + aR)/(gamma - 1))
dt = intersection_time * CFL

# Solving for intersection x position
intersection_pos = (uL + 2*aL/(gamma - 1)) * intersection_time
```

```python
plt.plot(x1L, x, color='navy', label='x1L')
plt.plot(x2L, x, color='orangered', label='x2L')
plt.plot(x1R, x, color='navy', linestyle='--', label='x1R')
plt.plot(x2R, x, color='orangered', linestyle='--', label='x2R')

plt.plot(intersection_pos, dt, 'o', color='green')

plt.ylim(x[0], x[-1])
plt.legend()
plt.show()
```

```python
# Values for f, or force term
Cd = 0.47
dia = 115e-6 # m

A = np.pi/4 * dia**2
first = True
```

```python
# Solve for f

# Set initial boundary values
P_left = P_arr[0][0]
P_right = P_arr[0][2]

u_left = u_arr[0][0]
u_right = u_arr[0][2]

T_left = T_arr[0][0]
T_right = T_arr[0][2]

rho_left = rho_arr[0][0]
rho_right = rho_arr[0][2]

# Drag force
dragL = 0.5*rho0*uL*uL*Cd*A
dragR = 0.5*rho0*uR*uR*Cd*A

# Sum of all forces (shown above)
fL = -dragL
fR = -dragR

# Solve for state variables at 0, 1, and 2 again.

# NOTE: Left and right boundary conditions are assumed to be the
# 0 and 2 state variable values at time t=0.

# 1. Find rho at the next time step.
rho0_new = rho0 + dt/dx*(rho_left*u_left - rho1*u1)
inv_rho0_new = 1/rho0_new

rho1_new = rho1 + dt/dx*(rho0*u0 - rho2*u2)
inv_rho1_new = 1/rho1_new

rho2_new = rho2 + dt/dx*(rho1*u1 - rho_right*u_right)
inv_rho2_new = 1/rho2_new

# 2. Find u at the next time step.
u0_new = inv_rho0_new * (rho0*u0 + dt/dx*((rho_left*u_left**2 + P_left) - (rho1*u1**2 + P1)))
u1_new = inv_rho1_new * (rho1*u1 + dt/dx*((rho0*u0**2 + P0) - (rho2*u2**2 + P2)))
u2_new = inv_rho2_new * (rho2*u2 + dt/dx*((rho1*u1**2 + P1) - (rho_right*u_right**2 + P_right)))

# 3. Find T at the next time step.
e0 = cp*T0
e1 = cp*T1
e2 = cp*T2

E0 = e0 + u0**2/2
E1 = e1 + u1**2/2
E2 = e2 + u2**2/2

h0 = P0/rho0
h1 = P1/rho1
h2 = P2/rho2

if first_2:
    h_left = h0
    h_right = h2
    first_2 = False

E0_new = inv_rho0_new * (rho0*E0 + dt/dx*((rho_left*u_left*(h_left + u_left**2/2)) - (rho1*u1*(h1 + u1**2/2))))
E1_new = inv_rho1_new * (rho1*E1 + dt/dx*((rho0*u0*(h0 + u0**2/2)) - (rho2*u2*(h2 + u2**2/2))))
E2_new = inv_rho2_new * (rho2*E2 + dt/dx*((rho1*u1*(h1 + u1**2/2)) - (rho_right*u_right*(h_right + u_right**2/2))))

T0_new = (E0_new - u0_new**2/2)/cp
T1_new = (E1_new - u1_new**2/2)/cp
T2_new = (E2_new - u2_new**2/2)/cp

# 4. Solve for P using equation of state.
P0_new = rho0_new*R*T0_new
P1_new = rho1_new*R*T1_new
P2_new = rho2_new*R*T2_new

# Save the data.
P_arr.append([P0_new, P1_new, P2_new])
u_arr.append([u0_new, u1_new, u2_new])
T_arr.append([T0_new, T1_new, T2_new])
rho_arr.append([rho0_new, rho1_new, rho2_new])
t_arr.append(t_arr[-1] + dt)
```

```python
steps = 5

first = True
for i in range(steps):
    # Set initial values at the interfaces with the cell values
    uL = (u0 + u1)/2
    TL = (T0 + T1)/2
    PL = (P0 + P1)/2
    rhoL = (rho0 + rho1)/2

    uR = (u1 + u2)/2
    TR = (T1 + T2)/2
    PR = (P1 + P2)/2
    rhoR = (rho1 + rho2)/2
    
    ################################################################################
    
    # Initial and populate time array
    t = np.arange(1, 51) * 0.000001 - 0.000001

    # Initial and populate space array
    x = np.arange(1, 51) * 0.000001

    # Speed of sound variables
    aL = np.sqrt(gamma*R*TL)
    print('TL: ', TL)
    print('aL: ', aL)
    print()
    aR = np.sqrt(gamma*R*TR)

    # Defining characteristic lines
    x1L = (uL + 2*aL/(gamma - 1)) * t
    x2L = (uL - 2*aL/(gamma - 1)) * t

    x1R = (uR + 2*aR/(gamma - 1)) * t + dx
    x2R = (uR - 2*aR/(gamma - 1)) * t + dx

    # Solving for intersection time, and multiplying by the CFL number
    intersection_time = dx/((uL - uR) + 2*(aL + aR)/(gamma - 1))
    dt = intersection_time * CFL

    # Solving for intersection x position
    intersection_pos = (uL + 2*aL/(gamma - 1)) * intersection_time
    
    ################################################################################
    
    # Solve for f
    
    if first:
        # Set initial boundary values
        P_left = P_arr[0][0]
        P_right = P_arr[0][2]

        u_left = u_arr[0][0]
        u_right = u_arr[0][2]

        T_left = T_arr[0][0]
        T_right = T_arr[0][2]

        rho_left = rho_arr[0][0]
        rho_right = rho_arr[0][2]
        
        first = False
        first_2 = True

    # Drag force
    dragL = 0.5*rho0*uL*uL*Cd*A
    dragR = 0.5*rho0*uR*uR*Cd*A

    # Sum of all forces (shown above)
    fL = -dragL
    fR = -dragR

    # Solve for state variables at 0, 1, and 2 again.

    # NOTE: Left and right boundary conditions are assumed to be the
    # 0 and 2 state variable values at time t=0.

    # 1. Find rho at the next time step.
    rho0_new = rho0 + dt/dx*(rho_left*u_left - rho1*u1)
    inv_rho0_new = 1/rho0_new

    rho1_new = rho1 + dt/dx*(rho0*u0 - rho2*u2)
    inv_rho1_new = 1/rho1_new

    rho2_new = rho2 + dt/dx*(rho1*u1 - rho_right*u_right)
    inv_rho2_new = 1/rho2_new

    # 2. Find u at the next time step.
    u0_new = inv_rho0_new * (rho0*u0 + dt/dx*((rho_left*u_left**2 + P_left) - (rho1*u1**2 + P1)))
    u1_new = inv_rho1_new * (rho1*u1 + dt/dx*((rho0*u0**2 + P0) - (rho2*u2**2 + P2)))
    u2_new = inv_rho2_new * (rho2*u2 + dt/dx*((rho1*u1**2 + P1) - (rho_right*u_right**2 + P_right)))

    # 3. Find T at the next time step.
    e0 = cp*T0
    e1 = cp*T1
    e2 = cp*T2

    E0 = e0 + u0**2/2
    E1 = e1 + u1**2/2
    E2 = e2 + u2**2/2

    h0 = P0/rho0
    h1 = P1/rho1
    h2 = P2/rho2

    if first_2:
        h_left = h0
        h_right = h2
        first_2 = False

    E0_new = inv_rho0_new * (rho0*E0 + dt/dx*((rho_left*u_left*(h_left + u_left**2/2)) - (rho1*u1*(h1 + u1**2/2))))
    E1_new = inv_rho1_new * (rho1*E1 + dt/dx*((rho0*u0*(h0 + u0**2/2)) - (rho2*u2*(h2 + u2**2/2))))
    E2_new = inv_rho2_new * (rho2*E2 + dt/dx*((rho1*u1*(h1 + u1**2/2)) - (rho_right*u_right*(h_right + u_right**2/2))))

    T0_new = (E0_new - u0_new**2/2)/cp
    T1_new = (E1_new - u1_new**2/2)/cp
    T2_new = (E2_new - u2_new**2/2)/cp

    # 4. Solve for P using equation of state.
    P0_new = rho0_new*R*T0_new
    P1_new = rho1_new*R*T1_new
    P2_new = rho2_new*R*T2_new

    # Save the data.
    P_arr.append([P0_new, P1_new, P2_new])
    u_arr.append([u0_new, u1_new, u2_new])
    T_arr.append([T0_new, T1_new, T2_new])
    rho_arr.append([rho0_new, rho1_new, rho2_new])
    t_arr.append(t_arr[-1] + dt)
    
    # Reset the variables.
    P0 = P0_new
    P1 = P1_new
    P2 = P2_new

    u0 = u0_new
    u1 = u1_new
    u2 = u2_new

    T0 = T0_new
    T1 = T1_new
    T2 = T2_new
    
    rho0 = rho0_new
    rho1 = rho1_new
    rho2 = rho2_new
```

```python
# Transpose the matrices.

P_arr_t = []
u_arr_t = []
T_arr_t = []
rho_arr_t = []
temp = []

for i in range(len(P_arr[0])):
    for j in range(len(P_arr)):
        temp.append(P_arr[j][i])
    P_arr_t.append(temp)
    temp = []

for i in range(len(u_arr[0])):
    for j in range(len(u_arr)):
        temp.append(u_arr[j][i])
    u_arr_t.append(temp)
    temp = []
    
for i in range(len(T_arr[0])):
    for j in range(len(T_arr)):
        temp.append(T_arr[j][i])
    T_arr_t.append(temp)
    temp = []
    
for i in range(len(rho_arr[0])):
    for j in range(len(rho_arr)):
        temp.append(rho_arr[j][i])
    rho_arr_t.append(temp)
    temp = []
```
