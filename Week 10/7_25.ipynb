```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import LineString
```

# Riemann Solver Exercise with 2 Cell Interfaces

```python
# Ideal gas constant for air
R = 287 # J/g*K

# Set initial values for each cell
u0 = 600 # m/s
T0 = 300 # K
P0 = 100 # kPa
rho0 = P0/(R*T0)

u1 = 500 # m/s
T1 = 200 # K
P1 = 200 # kPa
rho1 = P1/(R*T1)

u2 = 400 # m/s
T2 = 100 # K
P2 = 300 # kPa
rho2 = P2/(R*T2)

# Set initial values at the interfaces with the cell values
uL = (u0 + u1)/2
TL = (T0 + T1)/2
PL = (P0 + P1)/2
rhoL = (rho0 + rho1)/2

uR = (u1 + u2)/2
TR = (T1 + T2)/2
PR = (P1 + P2)/2
rhoR = (rho1 + rho2)/2

# Other constants
dx0 = 0.1 # m
CFL = 0.5
gamma = 1.4 # for air
```

```python
# Initial and populate time array
t = np.arange(1, 51) * 0.000001 - 0.000001
    
# Initial and populate space array
x = np.arange(1, 51) * 0.000001 - 0.0000025

# Speed of sound variables
aL = np.sqrt(gamma*R*TL)
aR = np.sqrt(gamma*R*TR)

# Defining characteristic lines
x1L = (uL + 2*aL/(gamma - 1)) * t
x2L = (uL - 2*aL/(gamma - 1)) * t

x1R = (uR + 2*aR/(gamma - 1)) * t + dx0
x2R = (uR - 2*aR/(gamma - 1)) * t + dx0

x
```

```python
plt.plot(x1L, x, color='navy', label='x1L')
plt.plot(x2L, x, color='orangered', label='x2L')
plt.plot(x1R, x, color='navy', linestyle='--', label='x1R')
plt.plot(x2R, x, color='orangered', linestyle='--', label='x2R')


plt.legend()
plt.show()
```
